```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/uber.csv")
```

```python
df_clean = df.copy()
```

```python
print(f"Initial rows: {len(df_clean)}")
print(f"Initial columns: {df_clean.shape[1]}")
```

```python
df_clean[df_clean.isnull().any(axis=1)]
```

```python
before = len(df_clean)
df_clean = df_clean.dropna()
after = len(df_clean)
print(f"Rows removed: {before-after}")
print(f"Remaining rows: {after}")
```

```python
df_clean[df_clean["fare_amount"] <= 0]
```

```python
before = len(df_clean)
df_clean = df_clean[df_clean["fare_amount"] > 0]
after = len(df_clean)
print(f"Rows removed: {before-after}")
print(f"Remaining rows: {after}")
```

```python
before = len(df_clean)
df_clean = df_clean[(df_clean["passenger_count"] > 0) & (df_clean["passenger_count"] < 7)]
after = len(df_clean)
print(f"Rows removed: {before-after}")
print(f"Remaining rows: {after}")
```

```python
invalid_coords = df_clean[
    (df_clean["pickup_longitude"] < -180) |
    (df_clean["pickup_longitude"] > 180) |
    (df_clean["pickup_latitude"] < -90) |
    (df_clean["pickup_latitude"] > 90) |
    (df_clean["dropoff_longitude"] < -180) |
    (df_clean["dropoff_longitude"] > 180) |
    (df_clean["dropoff_latitude"] < -90) |
    (df_clean["dropoff_latitude"] > 90)
]

invalid_coords
```

```python
before = len(df_clean)

df_clean = df_clean[
    (df_clean["pickup_longitude"].between(-180, 180)) &
    (df_clean["pickup_latitude"].between(-90, 90)) &
    (df_clean["dropoff_longitude"].between(-180, 180)) &
    (df_clean["dropoff_latitude"].between(-90, 90))
]

after = len(df_clean)

print(f"Rows removed: {before-after}")
print(f"Remaining rows: {after}")
```

```python
zero_coords = df_clean[
    (
        (df_clean["pickup_longitude"] == 0) &
        (df_clean["pickup_latitude"] == 0)
    ) |
    (
        (df_clean["dropoff_longitude"] == 0) &
        (df_clean["dropoff_latitude"] == 0)
    )
]

zero_coords
```

```python
before = len(df_clean)

df_clean = df_clean[
    ~(
        (
            (df_clean["pickup_longitude"] == 0) &
            (df_clean["pickup_latitude"] == 0)
        ) |
        (
            (df_clean["dropoff_longitude"] == 0) &
            (df_clean["dropoff_latitude"] == 0)
        )
    )
]

after = len(df_clean)

print(f"Rows removed: {before-after}")
print(f"Remaining rows: {after}")
```

```python
df_clean["pickup_datetime"] = pd.to_datetime(df_clean["pickup_datetime"])
```
